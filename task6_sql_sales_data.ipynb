```python
import pandas as pd
import sqlite3
from google.colab import files

# Step 1: Upload your Kaggle retail dataset CSV
print("Upload your retail sales dataset ")
uploaded = files.upload()

df = pd.read_csv(list(uploaded.keys())[0])
display(df.head())

# Step 2: Create SQLite in-memory DB
conn = sqlite3.connect(':memory:')
df.to_sql('retail_sales', conn, index=False, if_exists='replace')

# Step 3: Run SQL query
query = """
SELECT
    strftime('%Y', Date) AS year,
    strftime('%m', Date) AS month,
    SUM("Total Amount") AS total_revenue,
    COUNT(DISTINCT "Transaction ID") AS total_orders
FROM
    retail_sales
GROUP BY
    year, month
ORDER BY
    year ASC,
    month ASC
LIMIT 12;
"""
result = pd.read_sql_query(query, conn)
display(result)

#save the results
result.to_csv("task6_retail_sales_trend.csv", index=False)
files.download("task6_retail_sales_trend.csv")
```
